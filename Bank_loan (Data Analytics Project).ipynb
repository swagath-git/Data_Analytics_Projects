## Cleaning Dataset

```python
df = df.dropna()
```

```python
# Check missing values
df.isnull().sum()
```

```python
df['issue_date'].unique()[:10]
```

```python
df['issue_date'].dtype
```

```python
import warnings
warnings.simplefilter('ignore')

df['issue_date'] = pd.to_datetime(df['issue_date'], errors='coerce')
```

```python
df['month_year'] = df['issue_date'].dt.to_period('M')
```

```python
df['loan_status_category'] = df['loan_status'].apply(
    lambda x: 'Good' if x == 'Fully Paid' else 'Bad'
)
```
